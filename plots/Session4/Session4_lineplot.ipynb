### Title with line plot

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka plot")
plt.show
```

### Defining x and y limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show
```

### Set styles
- darkgrid
- White grid
- dark
- white
- ticks

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
phool = sns.load_dataset("iris")
phool

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka plot")

#style
sns.set_style(style=None, rc=None)
sns.set_style("dark")

#limits x and y
plt.xlim(2)
plt.ylim(1)
plt.show 
```

### Size of figures

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
phool = sns.load_dataset("iris")
phool

#figure size
plt.figure(figsize=(1,1))

#draw a line plot
sns.lineplot(x="sepal_length",y="sepal_width",data=phool)
plt.title("Phoolon ka plot")

#style
sns.set_style(style=None, rc=None)
sns.set_style("dark")

#limits x and y
plt.xlim(2)
plt.ylim(1)




plt.show 
```
