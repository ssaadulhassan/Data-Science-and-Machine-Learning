### Barplot on IRIS (Flower) data

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
phool = sns.load_dataset("iris")
phool

#draw a bar plot
sns.barplot(x="species",y="sepal_width",data=phool) # x categorical data ha   y is numeric data
plt.title("Phoolon ka plot")
plt.show
```

### Barplot on Titanic Data

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
kashti = sns.load_dataset("titanic")
kashti

#draw a bar plot
sns.barplot(x="sex",y="alone",hue="who",data=kashti) # x categorical data ha   y is numeric data
plt.title("Titanic ka bar plot")
plt.show
```

### Barplot on Titanic Data with Detailing

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#load data set
kashti = sns.load_dataset("titanic")
kashti

#figure size
plt.figure(figsize=(4,4))

#draw a bar plot
# order of data 
# ci graph se dande hatane k lye
#
sns.barplot(x="sex",y="alone",hue="who",data=kashti, order=["female","male"],color="brown",ci=None)
plt.title("Titanic ka plot")

#style
sns.set_style(style=None, rc=None)
sns.set_style("white")

#limits x and y
#plt.xlim(0)
#plt.ylim(0)




plt.show 
```

### Estimator usage

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from numpy import mean
#load data set
kashti = sns.load_dataset("titanic")
kashti

#figure size
plt.figure(figsize=(4,4))

#draw a line plot
# order of data 
# ci graph se dande hatane k lye (Confidence Interval)
# yahan tm ne order hataya tha tabhe sahe plot hua yad karlena tmhe lazmi yad aega
# color saturation
sns.barplot(x="class",y="fare",hue="sex",data=kashti,color="brown",ci=None,palette="pastel",estimator=mean,saturation=0.5)
plt.title("Titanic ka plot")

#style
sns.set_style(style=None, rc=None)
sns.set_style("white")

#limits x and y
#plt.xlim(0)
#plt.ylim(0)




plt.show 
```
